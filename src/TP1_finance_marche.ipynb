```python
import numpy as np
import matplotlib.pyplot as plt
```

**PARTIE 1**

```python
def f_X_inv(y):
    return (
        1 * (0 <= y <= 1 / 5)
        + 2 * (1 / 5 < y <= 7 / 10)
        + 3 * (7 / 10 < y <= 19 / 20)
        + 4 * (19 / 20 < y <= 1)
    )


def X():
    return f_X_inv(np.random.uniform())
```

```python
X_sample = []
for i in range(10000):
    X_sample.append(X())

counts, bins = np.histogram(X_sample, bins=4)
plt.hist(bins[:-1], bins, weights=counts / sum(counts))
plt.show()
```

```python
def f_Y_inv(y):
    lbda = 2
    return -np.log(1 - y) / lbda * (y > 0)


def Y():
    return f_Y_inv(np.random.uniform())
```

```python
Y_sample = []
for i in range(10000):
    Y_sample.append(Y())

counts, bins = np.histogram(Y_sample, bins=130)
plt.hist(bins[:-1], bins, weights=counts / sum(counts))
plt.show()
```

```python
Z_sample = [Y() ** X() for _ in range(100000)]

print(np.mean(Z_sample))
```

**PARTIE 2**

```python
T = 1
n = 1000
N = 5
step = T / n

B_data = []

for j in range(N):
    B = [0]
    for i in range(1, n + 1):
        B.append(B[i - 1] + np.sqrt(step) * np.random.normal())
    B_data.append(B)
```

```python
plt.figure()
for B in B_data:
    plt.plot(B)
plt.show()
```
